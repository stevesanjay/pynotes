```python
# Area Plot
```

```python
# import necessary library
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Data
```

```python
x = np.linspace(0, 10, 100)
```

```python
y = np.sin(x)
```

```python
# Creating an area plot
```

```python
plt.fill_between(x, y, color='skyblue', alpha=0.5)
```

```python
# Adding title and labels
```

```python
plt.title('Area Plot')
```

```python
plt.xlabel('X-axis')
```

```python
plt.ylabel('Y-axis')
```
