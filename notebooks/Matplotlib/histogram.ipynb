```python
# Histogram
```

```python
# import necessary library
```

```python
import matplotlib.pyplot as plt
```

```python
import numpy as np
```

```python
# Data
```

```python
data = np.random.randn(1000)
```

```python
# Creating the histogram
```

```python
plt.hist(data, bins=30, edgecolor='black')
```

```python
# Adding title and labels
```

```python
plt.title('Histogram')
```

```python
plt.xlabel('Value')
```

```python
plt.ylabel('Frequency')
```
