```python
# Basic Line Plot
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
```

```python
# Data
y = [2, 4, 6, 8, 10]
```

```python
# Creating the plot
plt.plot(x, y)
```

```python
# Adding title and labels
plt.title("Basic Line Plot")
```

```python
plt.xlabel("X-axis")
```

```python
plt.ylabel("Y-axis")
```

```python
# Show plot
plt.show()
```
