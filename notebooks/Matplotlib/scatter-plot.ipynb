```python
# Scatter Plot
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
x = [1, 2, 3, 4, 5]
```

```python
# Data
y = [10, 20, 25, 30, 40]
```

```python
# Creating the scatter plot
plt.scatter(x, y)
```

```python
# Adding title and labels
```

```python
plt.title('Scatter Plot')
```

```python
plt.xlabel('X-axis')
```

```python
plt.ylabel('Y-axis')
```

```python
plt.show()
```
