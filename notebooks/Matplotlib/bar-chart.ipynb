```python
# Bar Chart
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
```

```python
categories = ['A', 'B', 'C', 'D']
```

```python
values = [7, 12, 5, 18]
```

```python
# Creating the bar chart
```

```python
plt.bar(categories, values)
```

```python
# Adding title and labels
```

```python
plt.title('Bar Chart')
```

```python
plt.xlabel('Categories')
```

```python
plt.ylabel('Values')
```
