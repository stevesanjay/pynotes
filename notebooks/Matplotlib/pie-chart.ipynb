```python
# Pie Chart
```

```python
# import necessary library
```

```python
import matplotlib.pyplot as plt
```

```python
# Data
```

```python
sizes = [10, 20, 30, 40]
```

```python
labels = ['A', 'B', 'C', 'D']
```

```python
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue']
```

```python
# Creating the pie chart
```

```python
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
```

```python
# Adding title
```

```python
plt.title('Pie Chart')
```
